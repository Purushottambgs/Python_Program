```python
batsman=pd.read_csv('pandas/Dataset/sharma-kohli.csv')
batsman

plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.title('Rohit sharma vs Virat kohli career comparison')

plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='green')
plt.plot(batsman['index'], batsman['RG Sharma'], color='black')
plt.title('Rohit sharma vs Virat kohli career comparison')

plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed') # dashdot, 
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dotted')
plt.title('Rohit sharma vs Virat kohli career comparison')

plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='solid',linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot', linewidth=2)
plt.title('Rohit sharma vs Virat kohli career comparison')

plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', marker='D', markersize=10) 
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dotted', marker='+')
plt.title('Rohit sharma vs Virat kohli career comparison')

plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python

plt.plot(batsman['index'], batsman['V Kohli'], color='red', linewidth=4, linestyle='dashed')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', marker='D', markersize=5)
plt.title('Rohit and Virat ka score marging')

plt.xlabel('All Season')
plt.ylabel('Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', marker='D', markersize=10, label='virat') 
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dotted', marker='+', label='rohit')
plt.title('Rohit sharma vs Virat kohli career comparison')

plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best') # change kar sakte h
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', marker='D', markersize=10, label='virat') 
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dotted', marker='+', label='rohit')
plt.title('Rohit sharma vs Virat kohli career comparison')

plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
plt.grid()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', marker='D', markersize=10, label='virat') 
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dotted', marker='+', label='rohit')
plt.title('Rohit sharma vs Virat kohli career comparison')

plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
plt.grid()
plt.show()
```
